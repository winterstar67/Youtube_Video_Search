```python
Full_text = ""
for data in youtube_transcripts_data.snippets:
    Full_text += data.text + " " 
```

```python
# pip install wtpsplit regex nltk
from wtpsplit import SaT
import re

DISCOURSE = r"(?:okay|so|now|first of all|in this video|for example|in particular|since then|however|therefore|as a result)"
DISCOURSE_RE = re.compile(rf"\b{DISCOURSE}\b", re.IGNORECASE)

def segment(text: str, method="sat", threshold=0.3):
    txt = (text)
    seg = SaT("sat-1l-sm")  # 작고 빠른 모델
    raw = seg.split(txt, threshold=threshold)
    return (raw)

# 사용
sents = segment(Full_text, method="sat", threshold=0.35)
```

```python
# 맨 처음 split된 상태인 원본 데이터를 가지고 있는다
# 마찬가지로 원본 split의 start와 duration을 가지고 있는다
# 그리고 concat하고 split한 결과를 가져온다.

# 원본 split 데이터 각각을 좌측에서부터 align 시켜 나가면서 start와 duration을 확인한다

sents
```

```python
raw_index = 0
sents_with_metadata = []
for sentence in sents:
    print("=========== sentence ==========:", sentence)
    raw_index_at_sentence = raw_index
    start_at_sentence = 0
    for raw_split_string in youtube_transcripts_data[raw_index_at_sentence:]:
        print("raw_split_string:", raw_split_string)
        raw_index += 1
        
        if (start_at_sentence==0): # 텍스트의 시작 시간
            start_at_sentence = raw_split_string.start
        else:
            pass

        if (raw_split_string.text in sentence): # 진행 하면서 원본 텍스트가 SaT 텍스트에 포함되는 경우
            pass
        
        else:
            end_at_sentence = raw_split_string.start + raw_split_string.duration
            sents_with_metadata.append({"text":sentence, "start":start_at_sentence, "end":round(end_at_sentence,3)})
            break
```
